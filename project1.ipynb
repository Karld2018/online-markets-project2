```python
#my values for V_a and V_b
myvalue_a = 1.8
myvalue_b = 67.8

#my bids for Auction A and B
mybid_a = 0.9
mybid_b = 33.9

#Total number of students in class is 55
total_bids = 55

#To calculate probabilities of winning auctions A and B
prob_a = 0
prob_b = 0

for index,row in f1.iterrows():
    if(index != 47):
        if(row['bid in Auction A'] <= mybid_a):
            prob_a += 1
        if(row[' bid in Auction B'] <= mybid_b):
            prob_b += 1
        

    # print(row['v_A'],row['v_B'])
    # print(index)
    # count += 1
prob_a = prob_a/55
prob_b = prob_b/55   
print("Probability of me winning Auction A is:",prob_a)
print("Probability of me winning Auction B is:",prob_b)

util_a = prob_a*(myvalue_a - mybid_a)
util_b = prob_b*(myvalue_b - mybid_b)

print("Expected utility for me in Auction A is:",util_a)
print("Expected utility for me in Auction B is:",util_b)
```

```python
#probablility of winning an auction with a particular bid
def prob_win_a(bid,idx):
    prob = 0
    
    for index,row in f1.iterrows():
        if(index != idx):
            if(row['bid in Auction A'] <= bid):
                prob += 1
    return prob/55

def prob_win_b(bid,idx):
    prob = 0
    
    for index,row in f1.iterrows():
        if(index != idx):
            if(row[' bid in Auction B'] <= bid):
                prob += 1
    return prob/55
```

```python
optimal_a = 0
optimal_b = 0
best_bid = 0
for index,row in f1.iterrows():
    bid = row[' bid in Auction B']
    p = prob_win_b(bid,index)
    exp_val = p*(myvalue_b - bid)
    # print(exp_val)
    # print(bid)
    # print(myvalue_b - bid)
    # print(p)

    # print(mybid_b)
    if exp_val > optimal_b:
        optimal_b = exp_val
        best_bid = bid

print("Optimal expected value for Auction B is:",optimal_b)
print("Best bid is:",best_bid)
# print(p)
```
